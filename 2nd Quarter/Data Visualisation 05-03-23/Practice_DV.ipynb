```python
import numpy as np
import matplotlib.pyplot as plt
import sys
%matplotlib inline
```

```python
x = np.arange(1,101)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y, 'k--')
plt.xlabel('x')
plt.ylabel('sinx')
plt.title('Graph of sin x')
```

```python
x = np.linspace(1,10,100)
y = x+2
plt.plot(x,y, 'k--')
plt.xlabel('x')
plt.ylabel('sin x')
plt.grid(visible=True, which='major', color='b', linestyle='-')
plt.grid(visible=True, which='minor', color='r', linestyle='--')
plt.minorticks_on()
plt.title('Graph of Sin x')
```

```python
plt.plot(x,y, linestyle='dashed')
```

```python
plt.plot(x, np.tan(x))
plt.xlim(1,10)
plt.ylim(-1.5, 1.5)
```

### Visualising Multidimentional Data

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
features
```

```python
features.shape
```

```python
plt.scatter(features[0], features[1], alpha=0.3, s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
x = ['Viettel','VNPT','Mobiphone']
revenue = [37600000, 6445000, 6045000]
x_pos = [0,1,2]
plt.bar(x_pos, revenue, color='blue')
plt.xlabel('Telecom')
plt.ylabel('VND')
plt.title('Telecom service revenue in Vitenam')
plt.xticks(x_pos, x)
plt.show()
```
